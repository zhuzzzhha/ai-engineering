```python
import os
import torch
from torch.utils.data import Dataset, DataLoader
from PIL import Image
from stl import mesh
import numpy as np
from torchvision import transforms

class Sketch3DModelDataset(Dataset):
    def __init__(self, sketch_dir, stl_dir, transform=None):
        """
        Инициализация набора данных.
        :param sketch_dir: Путь к папке с эскизами.
        :param stl_dir: Путь к папке с STL файлами.
        :param transform: Трансформации для изображений.
        """
        self.sketch_dir = sketch_dir
        self.stl_dir = stl_dir
        self.file_indices = [f.split('.')[0] for f in os.listdir(stl_dir)]
        self.transform = transform

        self.max_triangles = self.find_max_triangles()

    def find_max_triangles(self):
        max_triangles = 0
        for i, stl_file in enumerate(os.listdir(self.stl_dir)):
            if i%100 == 0:
                print(stl_file)
            if stl_file.endswith('.stl'):
                stl_path = os.path.join(self.stl_dir, stl_file)
                try:
                    parsed_mesh = mesh.Mesh.from_file(stl_path)
                    num_triangles = parsed_mesh.vectors.shape[0]
                    max_triangles = max(max_triangles, num_triangles)
                except Exception as e:
                    print(f"Ошибка при обработке {stl_file}: {e}")
        print(max_triangles)
        return max_triangles

    def pad_tensor(self, tensor):
        """Дополняет тензор нулями до максимального размера."""
        padding_size = self.max_triangles - tensor.shape[0]
        if padding_size > 0:
            padding = torch.zeros((padding_size, tensor.shape[1], tensor.shape[2]), dtype=tensor.dtype)
            return torch.cat([tensor, padding], dim=0)
        return tensor

    def __len__(self):
        return len(self.file_indices)

    def __getitem__(self, idx):
        file_index = self.file_indices[idx]

        images = []
        for view in ['f', 'r', 't']:
            img_path = os.path.join(self.sketch_dir, f"{file_index}_{view}.png")
            if not os.path.exists(img_path):
                raise FileNotFoundError(f"Файл {img_path} не найден.")
            image = Image.open(img_path)
            images.append(image)


        width, height = images[0].size
        combined_image = Image.new('RGB', (width * 3, height))

        for i, img in enumerate(images):
            combined_image.paste(img, (i * width, 0))


        if self.transform:
            combined_image = self.transform(combined_image)
        else:
            combined_image = transforms.ToTensor()(combined_image)


        stl_path = os.path.join(self.stl_dir, f"{file_index}.stl")
        try:
            parsed_mesh = mesh.Mesh.from_file(stl_path)
            target_vectors = torch.tensor(parsed_mesh.vectors.copy(), dtype=torch.float32)
        except Exception as e:
            raise RuntimeError(f"Ошибка при загрузке STL файла {stl_path}: {e}")

        target_vectors = self.pad_tensor(target_vectors)

        return combined_image, target_vectors
```

```python
class Simplified3DModel(nn.Module):
    def __init__(self):
        super(Simplified3DModel, self).__init__()
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, stride=1, padding=1)
        self.pool = nn.MaxPool2d(2, 2)  # Снижает размерность в 2 раза
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)


        self.pool2 = nn.MaxPool2d(2, 2)

        self.flatten_size = 64 * (180 // 4) * (540 // 4)  

        self.fc1 = nn.Linear(self.flatten_size, 396 * 3 * 3)  
        self.fc2 = nn.Linear(396 * 3 * 3, 396 * 3 * 3)  

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool2(F.relu(self.conv2(x)))
        x = x.view(x.size(0), -1)  
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x.view(x.size(0), 396, 3, 3)  
```

```python

model = Simplified3DModel().to('cuda')
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
scaler = GradScaler()


num_epochs = 10
train_losses, val_losses = [], []
```

```python
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0

    for sketches, targets in train_loader:

        sketches, targets = sketches.to('cuda'), targets.to('cuda')

        optimizer.zero_grad()


        outputs = model(sketches)
        loss = criterion(outputs, targets)


        loss.backward()
        optimizer.step()

        running_loss += loss.item()


        del sketches, targets, outputs, loss
        gc.collect()

    train_loss = running_loss / len(train_loader)
    train_losses.append(train_loss)


    model.eval()
    val_loss = 0.0
    with torch.no_grad():
        for sketches, targets in val_loader:
            sketches, targets = sketches.to('cuda'), targets.to('cuda')
            outputs = model(sketches)
            loss = criterion(outputs, targets)
            val_loss += loss.item()

            del sketches, targets, outputs, loss
            gc.collect()

    val_loss /= len(val_loader)
    val_losses.append(val_loss)

    print(f"Epoch [{epoch+1}/{num_epochs}], Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}")
```

```python
import os
import numpy as np
from stl import mesh

def analyze_stl_files(stl_dir):
    num_triangles_list = []

    # Проходим по всем STL файлам в директории
    for stl_file in os.listdir(stl_dir):
        if stl_file.endswith('.stl'):
            stl_path = os.path.join(stl_dir, stl_file)
            try:
                parsed_mesh = mesh.Mesh.from_file(stl_path)
                num_triangles = parsed_mesh.vectors.shape[0]
                num_triangles_list.append((stl_file, num_triangles))
            except Exception as e:
                print(f"Ошибка при обработке {stl_file}: {e}")

    # Если нет STL файлов, выводим предупреждение
    if not num_triangles_list:
        print("Нет STL файлов для анализа.")
        return

    # Сортируем по количеству треугольников в порядке убывания
    num_triangles_list.sort(key=lambda x: x[1], reverse=True)

    # Топ-10 STL файлов по количеству треугольников
    print("Топ 10 STL файлов по количеству треугольников:")
    for i, (stl_file, num_triangles) in enumerate(num_triangles_list[:10]):
        print(f"{i+1}. {stl_file}: {num_triangles} треугольников")

    # Математические статистики
    num_triangles_arr = np.array([num for _, num in num_triangles_list])
    min_triangles = np.min(num_triangles_arr)
    mean_triangles = np.mean(num_triangles_arr)
    median_triangles = np.median(num_triangles_arr)

    print(f"\nМинимальное количество треугольников: {min_triangles}")
    print(f"Среднее количество треугольников: {mean_triangles:.2f}")
    print(f"Медианное количество треугольников: {median_triangles}")

    # Подсчет файлов с количеством треугольников <= среднему и медианному
    count_less_than_or_equal_mean = np.sum(num_triangles_arr <= mean_triangles)
    count_less_than_or_equal_median = np.sum(num_triangles_arr <= median_triangles)

    print(f"\nКоличество файлов с количеством треугольников <= среднему: {count_less_than_or_equal_mean}")
    print(f"Количество файлов с количеством треугольников <= медианному: {count_less_than_or_equal_median}")
```

```python
import os
import shutil
import numpy as np
from stl import mesh

def analyze_and_move_stl_files(stl_dir, output_dir):
    num_triangles_list = []
    files_to_move = []

    # Проходим по всем STL файлам в директории и находим количество треугольников
    for stl_file in os.listdir(stl_dir):
        if stl_file.endswith('.stl'):
            stl_path = os.path.join(stl_dir, stl_file)
            try:
                parsed_mesh = mesh.Mesh.from_file(stl_path)
                num_triangles = parsed_mesh.vectors.shape[0]
                num_triangles_list.append((stl_file, num_triangles))
            except Exception as e:
                print(f"Ошибка при обработке {stl_file}: {e}")

    if not num_triangles_list:
        print("Нет STL файлов для анализа.")
        return

    # Математические статистики
    num_triangles_arr = np.array([num for _, num in num_triangles_list])
    mean_triangles = np.mean(num_triangles_arr)

    # Фильтрация файлов, где количество треугольников <= среднему
    for stl_file, num_triangles in num_triangles_list:
        if num_triangles <= mean_triangles:
            files_to_move.append(stl_file)

    # Создаем папку для файлов, если она не существует
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Перемещаем соответствующие файлы в новую папку
    for stl_file in files_to_move:
        stl_path = os.path.join(stl_dir, stl_file)
        destination_path = os.path.join(output_dir, stl_file)
        try:
            shutil.move(stl_path, destination_path)
            print(f"Перемещен файл {stl_file} в {output_dir}")
        except Exception as e:
            print(f"Ошибка при перемещении {stl_file}: {e}")


# Папка с STL файлами
stl_dir = '/content/STL_Data'
# Папка для перемещения файлов, которые соответствуют условию
output_dir = 'FILTERED_STL_DATA'

# Запуск функции анализа и перемещения файлов
analyze_and_move_stl_files(stl_dir, output_dir)
```

```python
class Sketch3DModelDataset(Dataset):
    def __init__(self, sketch_dir, stl_dir, transform=None):
        self.sketch_dir = sketch_dir
        self.stl_dir = stl_dir
        self.file_indices = [f.split('.')[0] for f in os.listdir(stl_dir)]
        self.transform = transform
        self.max_triangles = self.find_max_triangles()

    def find_max_triangles(self):
        max_triangles = 0
        for i, stl_file in enumerate(os.listdir(self.stl_dir)):
            if i % 100 == 0:
                print(stl_file)
            if stl_file.endswith('.stl'):
                stl_path = os.path.join(self.stl_dir, stl_file)
                try:
                    parsed_mesh = mesh.Mesh.from_file(stl_path)
                    num_triangles = parsed_mesh.vectors.shape[0]
                    max_triangles = max(max_triangles, num_triangles)
                except Exception as e:
                    print(f"Ошибка при обработке {stl_file}: {e}")
        return max_triangles

    def pad_tensor(self, tensor):
        # Паддинг для количества треугольников до максимального
        padding_size = self.max_triangles - tensor.shape[0]
        if padding_size > 0:
            # Создаем тензор с нулями для добавления недостающих треугольников
            padding = torch.zeros((padding_size, 3, 3), dtype=tensor.dtype)
            return torch.cat([tensor, padding], dim=0)
        return tensor

    def __len__(self):
        return len(self.file_indices)

    def __getitem__(self, idx):
        file_index = self.file_indices[idx]
        images = []
        for view in ['f', 'r', 't']:
            img_path = os.path.join(self.sketch_dir, f"{file_index}_{view}.png")
            if not os.path.exists(img_path):
                raise FileNotFoundError(f"Файл {img_path} не найден.")
            image = Image.open(img_path)
            images.append(image)

        width, height = images[0].size
        combined_image = Image.new('RGB', (width * 3, height))
        for i, img in enumerate(images):
            combined_image.paste(img, (i * width, 0))

        if self.transform:
            combined_image = self.transform(combined_image)
        else:
            combined_image = transforms.ToTensor()(combined_image)

        stl_path = os.path.join(self.stl_dir, f"{file_index}.stl")
        try:
            parsed_mesh = mesh.Mesh.from_file(stl_path)
            target_vectors = torch.tensor(parsed_mesh.vectors.copy(), dtype=torch.float32)

            # Создаем двумерный массив для координат треугольников
            target_vectors = target_vectors.view(-1, 3, 3)  # [num_triangles, 3, 3] - каждый треугольник 3 вершины по 3 координаты

        except Exception as e:
            raise RuntimeError(f"Ошибка при загрузке STL файла {stl_path}: {e}")

        # Вывод размера до паддинга
        print(f"Before padding - Target shape for {file_index}: {target_vectors.shape}")

        # Паддинг до максимального количества треугольников
        target_vectors = self.pad_tensor(target_vectors)

        # Вывод размера после паддинга
        print(f"After padding - Target shape for {file_index}: {target_vectors.shape}")
        target_x = target_vectors[:, 0, :].reshape(1, -1, 1).squeeze(1)  # Преобразуем X в [1, 396*3, 1]
        target_y = target_vectors[:, 1, :].reshape(1, -1, 1).squeeze(1)  # Преобразуем Y в [1, 396*3, 1]
        target_z = target_vectors[:, 2, :].reshape(1, -1, 1).squeeze(1)  # Преобразуем Z в [1, 396*3, 1]

        return combined_image, target_x, target_y, target_z
```
